# performing hyper-parameter optimization in k-nn

```python
len(X_train)
```

```python
k=4
index=[i for i in range(len(X_train)) ]
index=set(index)
m=int(len(X_train)/k)
m
accuracy=[0 for i in range(4)]
final_accuracy=list()
accuracy
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
for l in range(1,30):
    clf=KNeighborsClassifier(n_neighbors=l)
    for i in range(0,k):
        cv_index=[j for j in range(i*m,i*m+m)]
        cv_index=set(cv_index)
        train_index=index-cv_index
        train_x=X_train.iloc[list(train_index)]
        cv_x=X_train.iloc[list(cv_index)]
        train_y=y_train.iloc[list(train_index)]
        cv_y=y_train.iloc[list(cv_index)]
        clf.fit(train_x,train_y)
        predict_y=clf.predict(cv_x)
        z=accuracy_score(cv_y,predict_y)
        accuracy[i]=z
    z=sum(accuracy)/4
    final_accuracy.append(z)
    
        
```

# Performing HyperParameter optimization in Logistic Regression

```python
k=4
index=[i for i in range(len(X_train))]
index=set(index)
m=int(len(X_train)/k)
m
accuracy=[0 for i in range(4)]
final_accuracy=list()
accuracy
n=[100,120,150,500,1000,1200,1500,2000,3000,3500,4000,5000,6000,7000,8000,9000,10000]
```

```python
for l in n:
    clf = LogisticRegression(random_state=0,C=l)
    for i in range(0,k):
        cv_index=[j for j in range(i*m,i*m+m)]
        cv_index=set(cv_index)
        train_index=index-cv_index
        train_x=X_train.iloc[list(train_index)]
        cv_x=X_train.iloc[list(cv_index)]
        train_y=y_train.iloc[list(train_index)]
        cv_y=y_train.iloc[list(cv_index)]
        clf.fit(train_x,train_y)
        predict_y=clf.predict(cv_x)
        z=accuracy_score(cv_y,predict_y)
        accuracy[i]=z
    z=sum(accuracy)/4
    final_accuracy.append(z)
    
        
```

```python
z=final_accuracy.index(max(final_accuracy))
print(n[z])
```

```python
plt.plot(n,final_accuracy)
plt.xlabel("C-value")
plt.ylabel("Accuracy")
plt.plot()
```

# Final Logistic Regression Model

```python
 clf = LogisticRegression(random_state=0,C=n[z],penalty='l2',solver='liblinear').fit(X_train,y_train)
```

```python
y_predict=clf.predict(X_test)
accuracy_score(y_test,y_predict)
```
